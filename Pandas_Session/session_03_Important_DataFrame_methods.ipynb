### `1. value_count()`
In pandas, `value_counts()` is a useful function in Pandas to count the occurrences of unique values in a Series (or columns of DataFrames). It returns a Series with the frequency of each unique value sorted in descending order.

```python
a = pd.Series([1,1,2,2,3,3,3,3])
a.value_counts()
```

```python
# value_counts(series and dataframe)
marks = pd.DataFrame([
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,70,14],
    [80,70,14]
],columns=['iq','marks','package'])

marks
```

```python
marks.value_counts()
```

```python
ipl = pd.read_csv('Datasets\\ipl-matches.csv')
ipl
```

#### find which player has won most potm -> in finals and qualifiers

```python
ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts()
```

#### Toss decision pie plot

```python
ipl['TossDecision'].value_counts().plot(kind='pie',autopct='%0.1f%%')
```

#### how many matches each team has played

```python
(ipl['Team1'].value_counts() + ipl['Team2'].value_counts()).sort_values(ascending=False)
```

### `2. sort_values()`
In Pandas, the `sort_values()` function allows you to sort a DataFrame or Series by values. This is useful when you need to order your data based on specific column(s) or perform ranking operations.

```python
x = pd.Series([12,14,1,56,89])
x
```

```python
x.sort_values()
```

```python
x.sort_values(ascending=False)
```

```python
movies = pd.read_csv('Datasets\\movies.csv')
movies.head(2)
```

```python
movies.sort_values('title_x')
```

```python
movies.sort_values('title_x',ascending=False)
```

```python
students = pd.DataFrame(
    {
        'name':['nitish','ankit','rupesh',np.nan,'mrityunjay',np.nan,'rishabh',np.nan,'aditya',np.nan],
        'college':['bit','iit','vit',np.nan,np.nan,'vlsi','ssit',np.nan,np.nan,'git'],
        'branch':['eee','it','cse',np.nan,'me','ce','civ','cse','bio',np.nan],
        'cgpa':[6.66,8.25,6.41,np.nan,5.6,9.0,7.4,10,7.4,np.nan],
        'package':[4,5,6,np.nan,6,7,8,9,np.nan,np.nan]

    }
)
students
```

### `3. rank()`
In Pandas, the `rank()` function assigns ranks to elements in a Series or DataFrame.

```python
batsman = pd.read_csv('Datasets\\batsman_runs_ipl.csv')
batsman.head()
```

```python
batsman['batsman_rank'] = batsman['batsman_run'].rank(ascending=False)
batsman
```

```python
batsman.sort_values('batsman_rank')
```

### `4. sort_index()`
In Pandas, the `sort_index()` function is used to sort the rows (or columns) of a DataFrame or Series by their index values. 

```python
marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}
marks_series = pd.Series(marks)
marks_series
```

```python
marks_series.sort_index()
```

```python
marks_series.sort_index(ascending=False)
```

```python
movies.sort_index(ascending=False)
```
