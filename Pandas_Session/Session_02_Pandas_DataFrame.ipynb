# What is Pandas Dataframe?
A Data frame is a two-dimensional data structure or object, i.e., data is aligned in a tabular fashion in rows and columns. Pandas DataFrame consists of three principal components, the data, rows, and columns.


## Creating a DataFrame
- Pandas DataFrame can be created from the lists, dictionary, and from a list of dictionary etc.
- In the real world, a Pandas DataFrame will be created by loading the datasets from existing storage, storage can be SQL Database, CSV file, and Excel file. 

### DataFrame Using List

```python
students_data = [
    [100,80,10],
    [80,90,7],
    [90,60,7],
    [80,80,5]
]
pd.DataFrame(students_data, columns=['iq','Marks','Package'])
```

### Q.7 write a function that can return the track record of 2 teams against each other 

```python
def track_record(team1, team2):
  temp_df = ipl[((ipl['Team1'] == team1) & (ipl['Team2'] == team2)) | ((ipl['Team1'] == team2) & (ipl['Team2'] == team1))]
  total_matches = temp_df.shape[0]
  team1_wins = temp_df[temp_df['WinningTeam'] == team1].shape[0]
  team2_wins = temp_df[temp_df['WinningTeam'] == team2].shape[0]
  final_winner1 = ipl[(ipl['MatchNumber'] == 'Final') & (ipl['WinningTeam'] == team1)].shape[0]
  final_winner2 = ipl[(ipl['MatchNumber'] == 'Final') & (ipl['WinningTeam'] == team2)].shape[0]
  print(f'{team1} vs {team2} : {total_matches} matches')
  print(f'{team1} won : {team1_wins}')
  print(f'{team2} won : {team2_wins}')
  print(f'{team1} Won Finals {final_winner1}')
  print(f'{team2} Won Finals {final_winner2}')

track_record('Chennai Super Kings', 'Mumbai Indians')
```
